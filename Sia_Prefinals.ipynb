```python
import re

def parse_arff(file_path):
    with open(file_path, 'r') as file:
        data = file.read()

    # Extract relation name
    relation_match = re.search(r'@relation\s+(.*?)\s', data)
    relation_name = relation_match.group(1) if relation_match else None

    # Extract attribute names and types
    attribute_matches = re.findall(r'@attribute\s+(.*?)\s+(.*?)\s', data)
    attributes = [(name, type_) for name, type_ in attribute_matches]

    # Extract data instances
    data_match = re.search(r'@data\s+(.*?)\s', data, re.DOTALL)
    data_instances = data_match.group(1).strip().split('\n') if data_match else []

    return relation_name, attributes, data_instances

print(parse_arff("V4 data/2017 Q1.arff"))
```
